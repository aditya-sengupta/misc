```python
import numpy as np
from scipy.io import wavfile
from os import path
from pydub import AudioSegment                                                                 
src = "hamilton.mp3"
dst = "hamilton.wav"                                                       
sound = AudioSegment.from_mp3(src)
sound.export(dst, format="wav")

rate, tpp = wavfile.read(dst)
```

```python
rate
```

```python
tpp = tpp.flatten()
```

```python
tpp.shape[0]/rate
```

```python
tpp_freq = np.fft.fft(tpp)
```

```python
tpp_freq
```

```python
import matplotlib.pyplot as plt
plt.scatter(tpp_freq.real, tpp_freq.imag)
```

```python
plt.show()
```
